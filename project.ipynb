```python
import plotly.express as px
avg_sale_prices_neighborhood_fig = px.bar(df_train.reset_index(), 
                                          x='Neighborhood', y='SalePrice', 
                                          title="Average Sale Prices across Neighborhoods")
avg_sale_prices_neighborhood_fig.show()
```

```python
scatter_fig = px.scatter(df_train, x='GrLivArea', y='SalePrice', 
                         title='Living Area vs. Sale Price',
                         labels={"GrLivArea": "Ground Living Area (sq ft)", "SalePrice": "Sale Price"})
scatter_fig.show()
```

```python
hist_fig = px.histogram(df_train, x='SalePrice', 
                        title='Distribution of Sale Prices',
                        labels={"SalePrice": "Sale Price"},
                        nbins=100)
hist_fig.show()
```

```python
bubble_chart_fig = px.scatter(df_train, x='TotRmsAbvGrd', y='OverallQual', size='SalePrice', #or Bath_total
                              title="Rooms, Quality vs Sale Price",
                              labels={"TotRmsAbvGrd": "Total Rooms Above Grade", "OverallQual": "Overall Quality", "SalePrice": "Sale Price"},
                              size_max=60)


bubble_chart_fig.show()
```

```python
avg_sale_price_per_year = df_train.groupby('YrSold')['SalePrice'].mean().reset_index()

# Now, create the line chart with the average sale price
yearly_trends_fig = px.line(avg_sale_price_per_year, x='YrSold', y='SalePrice',
                            title='Yearly Trends in Sale Prices',
                            labels={"YrSold": "Year Sold", "SalePrice": "Average Sale Price"},
                            markers=True)
yearly_trends_fig.show()
```
